```python
xbar = df['Engine Size(L)'].mean()
ybar = df['CO2 Emissions(g/km)'].mean()
```

```python
print(xbar,ybar)
```

```python
##  sigma(x-xBar)*(y-yBar) / sigma(x-xBar)**2
```

```python
df["numerator"]=((df["Engine Size(L)"]-xbar)*(df["CO2 Emissions(g/km)"]-ybar)).sum()
df["denominator"]=((df["Engine Size(L)"]-xbar)**2).sum()
df
```

```python
theta1=df["numerator"]/df["denominator"]
theta1
```
