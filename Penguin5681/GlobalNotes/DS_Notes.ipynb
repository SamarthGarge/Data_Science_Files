## Shallow Copy
### Shallow Copy => creates a new array object, but the data elements are still references to the same elements in the original array
### in simple words: modifications made in the original array or in the shallow copy will affect each other
### view() => returns the shallow copy

```python
original_array = np.array([1, 2, 3, 4, 5])

shallow_copy = original_array.view()

# Modification in the shallow cp[y]
shallow_copy[0] = 99

print("Original array after modifying shallow copy:")
print(original_array)

print("Shallow copy after modification:")
print(shallow_copy)
```

## Deep Copy
### Deep Copy => A deep copy creates a new array object and also copies the data elements, meaning that the new array is completely independent of the original array.
### in simple words: once a deep copy is creating, making modifications in the original array will not affect the copy (vice versa is true)
### copy() => returns the deep copy

```python
original_array = np.array([1, 2, 3, 4, 5])

deep_copy = original_array.copy()

# modification in the deep copy
deep_copy[0] = 99

print("Original array after modifying deep copy:")
print(original_array)

print("Deep copy after modification:")
print(deep_copy)
```

## Working with some IRL Math equations: 

### Important Equations:

### Regression Coefficient: $Y = \theta_1x_1 + \theta_0$
### $\theta_1 = \frac{\Sigma((x - \bar x) * (y - \bar y))}{\Sigma (x - \bar x)^2}$

```python
# Calculating xBar and yBar (mean => average)
xBar = dataFrame['Engine Size(L)'].mean()
yBar = dataFrame['CO2 Emissions(g/km)'].mean()
```

```python
xBar, yBar
```

### Calculating $\theta_1$

```python
dataFrame['numerator'] = (dataFrame['Engine Size(L)'] - xBar) * (dataFrame['CO2 Emissions(g/km)'] - yBar)

dataFrame['denominator'] = ((dataFrame['Engine Size(L)'] - xBar) ** 2)
```

```python
thetaOne = dataFrame['numerator'].sum() / dataFrame['denominator'].sum()
```
