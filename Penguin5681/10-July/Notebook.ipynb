```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\prana\PycharmProjects\Data_Science_Files\Datasets\churn.csv');
df
```

```python
df.replace(0, np.nan, inplace=True)
```

```python
text = []
for i in df.State:
      text.append(i)

text
textStr = ''

for txt in text:
    textStr = textStr + str(txt) + ", "
    
textStr
```
