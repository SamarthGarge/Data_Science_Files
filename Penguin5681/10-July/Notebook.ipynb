```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\prana\PycharmProjects\Data_Science_Files\Datasets\churn.csv')
df
```

```python
# Replaces all occurences of 0 with nan

df.replace(0, np.nan, inplace=True)
```

```python
#  Collect all state values into a list and join them into a single string with each value separated by a comma

textStr = ' '.join(map(str, df['State']))

textStr
```

```python
obj = WordCloud().generate(text=textStr)
plt.imshow(obj, interpolation='bilinear')
```
